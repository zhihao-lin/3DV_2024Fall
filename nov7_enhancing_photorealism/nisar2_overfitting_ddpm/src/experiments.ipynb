```python
from src import seed_all
from src import run, gif_maker, extract_losses_from_checkpoints
from matplotlib import pyplot as plt
import os
seed_all(123)
```

# Experiment 1

```python
exp_name = 'exp1_copy'

ims = ['exp11', 'exp12', 'exp13']
for im in ims:
    root_dir = '/home/nisar2/Documents/cs598_3dv_hacker/hypersim_data'
    scenes_cams_frames_pth = f'/home/nisar2/Documents/cs598_3dv_hacker/src/{ims}.csv'
    output_pth = f'/home/nisar2/Documents/cs598_3dv_hacker/{exp_name}/{im}'
    run(root_dir, scenes_cams_frames_pth, output_pth)
    checkpoint_pth = f'/home/nisar2/Documents/cs598_3dv_hacker/{exp_name}/{im}/checkpoints/epoch_2000.pth'
    gif_maker(root_dir, scenes_cams_frames_pth, output_pth, checkpoint_pth, 0)


checkpoints_dir = [f'/home/nisar2/Documents/cs598_3dv_hacker/{exp_name}/{im}/checkpoints' for im in ims]

fig, ax = plt.subplots()
for i, checkpoint_dir in enumerate(checkpoints_dir):
    # Extract the loss values
    losses_df = extract_losses_from_checkpoints(checkpoint_dir)
    ax.plot(losses_df['epoch'], losses_df['loss'], marker='o', label=ims[i])
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.title("Loss per Epoch")

plt.legend(loc='upper right')
plt.show()
```
